## 1. Normalização Min-Max 📊
Na preparação de dados para Redes Neurais, é muito comum precisarmos colocar todas as variáveis numéricas na mesma escala (geralmente entre 0 e 1). A fórmula da Normalização Min-Max é:
$$X_{norm} = \frac{X - X_{min}}{X_{max} - X_{min}}$$

**Tarefa:** Crie um programa que faça a leitura (via teclado) de cinco valores. Calcule e exiba o valor normalizado de cada um deles.


```python
#usei o deepseek para auxiliar principalmente na sintaxe e um pouco na logica (estou acostumando com o python)

def main():
    valores = []
    print("Digite cinco números (um por linha ou separados por espaço):")

    while len(valores) < 5:
        try:
            linha = input().strip()
            if not linha:
                continue
            partes = linha.split()
            for p in partes:
                if len(valores) < 5:
                    num = float(p)
                    valores.append(num)
                else:
                    break
        except ValueError:
            print("Valor inválido. Digite apenas números.")

    min_val = min(valores)
    max_val = max(valores)

    if max_val == min_val:
        print("\nTodos os valores são iguais. Não é possível normalizar com Min-Max.")
        print("Valores normalizados definidos como 0.5 para todos.")
        norm_values = [0.5] * 5
    else:
        norm_values = [(x - min_val) / (max_val - min_val) for x in valores]

    print("\nResultados da normalização:")
    for i, (original, norm) in enumerate(zip(valores, norm_values), 1):
        print(f"Valor {i}: original = {original:.4f}  ->  normalizado = {norm:.4f}")

if __name__ == "__main__":
    main()
```

## 2. A Matriz de Confusão 🎲
Vamos avaliar um classificador binário gerado aleatoriamente!
1. Importe o módulo `random`.
2. Construa duas listas de 20 posições cada: `y_real` e `y_predito`. Preencha-as com números binários aleatórios (0 ou 1) usando `random.randint(0, 1)`.

**Tarefa:** Crie um dicionário para rastrear a matriz de confusão desse classificador.

**Matriz de Confusão:**
| | Real 0 | Real 1 |
| :--- | :--- | :--- |
| **Predito 0** | VN | FP |
| **Predito 1** | FN | VP |


```python
#usei o deepseek para auxiliar principalmente na sintaxe e um pouco na logica (estou acostumando com o python)

import random

# Gera listas aleatórias
y_real = [random.randint(0, 1) for _ in range(20)]
y_predito = [random.randint(0, 1) for _ in range(20)]

# Inicializa dicionário da matriz de confusão
confusao = {'VN': 0, 'FP': 0, 'FN': 0, 'VP': 0}

# Preenche a matriz
for real, pred in zip(y_real, y_predito):
    if real == 0 and pred == 0:
        confusao['VN'] += 1
    elif real == 0 and pred == 1:
        confusao['FP'] += 1
    elif real == 1 and pred == 0:
        confusao['FN'] += 1
    else:  # real == 1 and pred == 1
        confusao['VP'] += 1

# Exibe resultados
print("y_real    :", y_real)
print("y_predito :", y_predito)
print("\nMatriz de Confusão:")
print(f"VN (Verdadeiro Negativo): {confusao['VN']}")
print(f"FP (Falso Positivo)     : {confusao['FP']}")
print(f"FN (Falso Negativo)     : {confusao['FN']}")
print(f"VP (Verdadeiro Positivo): {confusao['VP']}")
```

## 3. Métricas Customizadas e Docstrings 📏
Usando o dicionário da Matriz de Confusão gerado no exercício anterior, vamos calcular o **F1-Score**.

O F1-Score é a média harmônica entre a *Precisão* e o *Recall*:
- **Precisão:** $\frac{VP}{VP + FP}$ (Dos que o modelo disse ser 1, quantos eram 1 de verdade?)
- **Recall:** $\frac{VP}{VP + FN}$ (Dos que eram realmente 1, quantos o modelo identificou?)
- **F1-Score:** $2 \times \frac{Precisão \times Recall}{Precisão + Recall}$

**Tarefa:**
1. Crie uma função chamada `calcular_f1_score(matriz_confusao: dict) -> float`.
2. Escreva uma **Docstring** completa para a função, descrevendo seu argumento e retorno.
3. Dentro da função, extraia os valores (VP, FP e FN) do dicionário recebido. Calcule a Precisão, o Recall e finalmente o F1-Score, retornando o valor final. *(Dica: Trate o caso de divisão por zero retornando 0.0 caso ocorra)*.
4. Ao final, chame a sua função passando o dicionário do Exercício 2 e exiba o resultado no console com 4 casas decimais.


```python
#usei o deepseek para auxiliar principalmente na sintaxe e um pouco na logica (estou acostumando com o python)

import random

y_real = [random.randint(0, 1) for _ in range(20)]
y_predito = [random.randint(0, 1) for _ in range(20)]

confusao = {'VN': 0, 'FP': 0, 'FN': 0, 'VP': 0}
for real, pred in zip(y_real, y_predito):
    if real == 0 and pred == 0:
        confusao['VN'] += 1
    elif real == 0 and pred == 1:
        confusao['FP'] += 1
    elif real == 1 and pred == 0:
        confusao['FN'] += 1
    else:
        confusao['VP'] += 1

def calcular_f1_score(matriz_confusao: dict) -> float:
    """
    Calcula o F1-Score a partir da matriz de confusão.
    Retorna 0.0 em casos de divisão por zero.
    """
    vp = matriz_confusao.get('VP', 0)
    fp = matriz_confusao.get('FP', 0)
    fn = matriz_confusao.get('FN', 0)

    precisao = vp / (vp + fp) if (vp + fp) != 0 else 0.0
    recall = vp / (vp + fn) if (vp + fn) != 0 else 0.0

    return 2 * (precisao * recall) / (precisao + recall) if (precisao + recall) != 0 else 0.0

f1 = calcular_f1_score(confusao)
print(f"F1-Score: {f1:.4f}")
```

## 4. Árvore Binária de Busca 🌳
**Tarefa:** Crie uma classe `NoArvore`. Ela deve receber no método construtor `__init__` um valor inteiro e inicializar as referências `esquerda` e `direita`. Crie dois métodos para essa classe:
- `inserir(self, valor: int)`: Se o valor for menor que o do nó atual, desce para a esquerda (se estiver livre, aloca um nó novo, se não, chama a inserção recursiva à esquerda). Faz o espelho para a direita.
- `buscar(self, valor: int) -> bool`: Retorna `True` se o valor existir na árvore sob esse nó, e `False` caso contrário, usando a lógica clássica da BST.


```python
#usei o deepseek para auxiliar principalmente na sintaxe e um pouco na logica (estou acostumando com o python)

class NoArvore:
    def __init__(self, valor: int):
        self.valor = valor
        self.esquerda = None
        self.direita = None

    def inserir(self, valor: int):
        if valor < self.valor:
            if self.esquerda is None:
                self.esquerda = NoArvore(valor)
            else:
                self.esquerda.inserir(valor)
        elif valor > self.valor:
            if self.direita is None:
                self.direita = NoArvore(valor)
            else:
                self.direita.inserir(valor)

    def buscar(self, valor: int) -> bool:
        if valor == self.valor:
            return True
        elif valor < self.valor and self.esquerda:
            return self.esquerda.buscar(valor)
        elif valor > self.valor and self.direita:
            return self.direita.buscar(valor)
        return False
```
